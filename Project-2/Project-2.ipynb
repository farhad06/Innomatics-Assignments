# Hypothesis Testing
### Using Chi-Square Testing


```python
df.groupby(['gender','occupation']).count()['age']
```

### Relationship beween occupation and gender

```python
crtab = pd.crosstab(df['occupation'],df['gender'])
crtab
```

```python
test_statistic,p,dof,_=stats.chi2_contingency(crtab)
p
```

```python
if p <0.05:
 print('There is no relation between the two variable')
else:
 print('There is some significance relation between the two variables')
```

### Relationship between gender and income

```python
crtab = pd.crosstab(df['occupation'],df['gender'])
crtab
```

```python
test_statistic,p,dof,_=stats.chi2_contingency(crtab)
p
```
